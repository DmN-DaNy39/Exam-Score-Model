```python
#initialising my libraries
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
#initialising my dataset
dataset = pd.read_csv('StudentPerformanceFactors.csv')
```

```python
#removing missing values
dataset.dropna()
```

```python
#setting what is to be predicted by the model
y = dataset.Exam_Score
```

```python
#setting what is inputted to the model to make a prediction
features = ['Attendance', 'Hours_Studied', 'Sleep_Hours', 'Physical_Activity']

X = dataset[features]
```

```python
#splitting the data for training and testing
train_X, val_X, train_y, val_y = train_test = train_test_split(X, y, random_state=1)
```

```python
#making the model
model = RandomForestRegressor(random_state=1)

#train the model with the test data
model.fit(train_X, train_y)
```

```python
#make a prediction based on the test data
predictions = model.predict(val_X)

#check how accurate the model is
print(mean_absolute_error(predictions, val_y))
```

```python
#Making the program more user friendly
#asking the user necessary questions
print("Please enter the following student performance factors:")
attendance = float(input("Attendance Percentage: "))
hours_studied = float(input("Hours of Studied Per Week: "))
sleep_hours = float(input("Hours of Sleep Per Night: "))
physical_activity = float(input("Hours of Physical Activity Per Week: "))

#putting user input into dataframe
user_input = pd.DataFrame([{
    'Attendance': attendance,
    'Hours_Studied': hours_studied,
    'Sleep_Hours': sleep_hours,
    'Physical_Activity': physical_activity
}])

#making a prediction from the user input
prediction = model.predict(user_input)

#giving the user a prediction
print(f"\nBased on the provided factors, the model has predicted your exam Score is: {prediction[0]:.2f}")
```
